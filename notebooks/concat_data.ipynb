```python
import pandas as pd
print(pd.__version__)
```

```python
df_list = [pd.DataFrame(i) for i in [scraper.books_scraper(), scraper.kingstone_scraper(), scraper.sanmin_scraper(), scraper.cite_scraper(), scraper.taaze_scraper(), scraper.tcsb_scraper()]]
df = pd.concat(df_list, axis=0, ignore_index=True)
```

```python
import sqlite3
```

```python
conn = sqlite3.connect("../data/03_processed/top_sales_books.db")
curs = conn.cursor()
curs.execute("drop table topsales")
curs.execute("create table topsales (Store, Rank, Name, Price, Link)")
for i in range(len(df)):
    curs.execute("insert into topsales values (?, ?, ?, ?, ?)", df.iloc[i])

conn.commit()
conn.close()
```
