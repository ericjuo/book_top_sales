```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("../data/03_processed/top_sales_books.db")
df_books = pd.read_sql("select rank, name from topsales where store == 'books'", conn)

df_books
```

```python
df_kingstone = pd.read_sql("select rank, name from topsales where store == 'kingstone'", conn)

df_kingstone
```

```python
df_sanmin = pd.read_sql("select rank, name from topsales where store == 'sanmin'", conn)

df_sanmin
```

```python
df_taaze = pd.read_sql("select rank, name from topsales where store == 'taaze'", conn)

df_taaze
```

```python
df_cite = pd.read_sql("select rank, name from topsales where store == 'cite'", conn)

df_cite
```

```python
df_tcsb = pd.read_sql("select rank, name from topsales where store == 'tcsb'", conn)

df_tcsb
```

```python
columns = ['TOP', '博客來', '金石堂', '三民書局', '讀冊', '城邦讀書花園', '墊腳石']
df_rank = pd.Series([i for i in range(1, 11)])
df = pd.concat([df_rank, df_books['Name'], df_kingstone['Name'], df_sanmin['Name'], df_taaze['Name'], df_cite['Name'], df_tcsb['Name']], axis=1)
df.columns = columns
df
```

```python
from datetime import date
today = date.today()
if today.month -1 == 0:
    rank_month = 12
    rank_year = today.year -1
else:
    rank_month = today.month -1
    rank_year = today.year

with open("../README.md", 'w') as f:
    f.write("# 台灣暢銷書排行榜  \n")
    f.write("## %s年%s月排行榜  \n" % (rank_year, rank_month))
    f.write("Updated on %s  \n" % today.strftime("%Y-%m-%d"))
    f.write(df.to_markdown(index=False))
```
